```python
import errno
import json
import os
import cv2

from scipy.ndimage import rotate
from scipy.stats import bernoulli

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.misc

import data_prep as dp



steering_angle=-0.5054564
Image_path=dp.IMG_PATH+"right_2017_12_17_20_41_21_579.jpg"
print("image path="+ Image_path)
img=plt.imread(Image_path)
plt.imshow(img)
plt.imsave('wr_image.jpg',img)
plt.show()
print("stering angle=",steering_angle)


print("\nShear image")
shear_image, steering_angle = dp.random_shear(img, steering_angle,100)
plt.imshow(shear_image)
plt.imsave('wr_shear_image.jpg',shear_image)
plt.show()
print("stering angle=",steering_angle)


cropped_image = dp.resize_image(shear_image, (64,64))
print("\nCropped image  new shape=",cropped_image.shape)
plt.imshow(cropped_image)
plt.imsave('wr_cropped_image.jpg',cropped_image)
plt.show()


print("\nflip_image")
fliped_image, steering_angle = dp.flip_image(cropped_image, steering_angle)
plt.imshow(fliped_image)
plt.imsave('wr_flip_image.jpg',fliped_image)
plt.show()
print("stering angle=",steering_angle)

print("\nRandom Bright image")
bright_image=dp.random_brightness(fliped_image)
plt.imshow(bright_image)
plt.imsave('wr_bright_image.jpg',bright_image)
plt.show()










```
